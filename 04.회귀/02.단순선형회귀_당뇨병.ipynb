## 단순 선형회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- 당뇨병 데이터셋

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target']= diabetes.target
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(18,9))
for i, feature in enumerate(df.columns[:-1]):
    row, col=i//5, i%5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- BMI vs target, BP vs target

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# BMI vs target
lr.fit(df.bmi.values.reshape(-1,1), diabetes.target)
lr.score(df.bmi.values.reshape(-1,1), diabetes.target)
```

```python
# BP vs target
lr.fit(df.bp.values.reshape(-1,1), diabetes.target)
lr.score(df.bp.values.reshape(-1,1), diabetes.target)
```

```python
print(diabetes.DESCR)
```

- 훈련/ 테스트 데이터셋 분리 (test_size=0.1)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.bmi.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2021
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 계수, 절편, 어쩌구
lr.coef_, 
```

```python
from sklearn.metrics import r2_score # 스코어로 계산했던 값
pred = lr.predict(X_test)
r2_score(y_test, pred)
```

```python
lr.score(X_test, y_test) # 두 결과값이 같다.
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
mse, rmse
```

- 시각화

```python
y_test[0], pred[0]
```

```python
plt.scatter(X_train, y_train)
plt.grid()
plt.xlabel('BMI')
plt.ylabel('target')
plt.title('BMI vs target', fontsize=15)
plt.show()
```

```python
weight, bias = np.polyfit(df.bmi, diabetes.target, 1) 
weight, bias
```

```python
xs = np.array([-0.1, 0.2])
ys = xs * weight + bias
ys
```

```python
plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test, label='test')
plt.scatter(X_test, pred, label='pred')
plt.plot(xs, ys, 'red', lw=3)
plt.grid()
plt.xlabel('BMI')
plt.ylabel('target')
plt.title('BMI vs target', fontsize=15)
plt.legend()
plt.show()
```

- BP vs target

```python
from sklearn.model_selection import train_test_split
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    df.bp.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2021
)
X_train2.shape, X_test2.shape, y_train2.shape, y_test2.shape
```

```python
lr = LinearRegression()
lr.fit(X_train2, y_train2)
```

```python
lr.coef_, lr.intercept_
```

```python
pred = lr.predict(X_test2)
r2_score(y_test2, pred)
```

```python
plt.scatter(X_train2, y_train2)
plt.grid()
plt.xlabel('BP')
plt.ylabel('target')
plt.title('BP vs target', fontsize=15)
plt.show()
```

```python
weight, bias = np.polyfit(df.bp, diabetes.target, 1) 
weight, bias
```

```python
xs = np.array([-0.1, 0.15])
ys = xs * weight + bias # xs * lr2.coef_ + lr2.intercept_
ys
```

```python
plt.scatter(X_train2, y_train2)
plt.scatter(X_test2, y_test2, label='test')
plt.scatter(X_test2, pred, label='pred')
plt.plot(xs, ys, 'red', lw=3)
plt.grid()
plt.xlabel('BP')
plt.ylabel('target')
plt.title('BP vs target', fontsize=15)
plt.legend()
plt.show()
```
